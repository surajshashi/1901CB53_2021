```python
import pandas as pd
raw_df=pd.read_html('https://raw.githubusercontent.com/diazoniclabs/Microland/main/onion_2019.html')
raw_df
```

```python
df=raw_df[-1]
```

```python
df=df[:-1]
```

```python
df
```

```python
pd.to_datetime(df['Date'])
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df=df.rename(columns={'Modal Price (Rs/q)':'Avg'})
```

```python
df
```

```python
df=df[['Date','Market','Avg']]
```

```python
df
```

```python
df['Avg']=df['Avg'].astype(int)
df
```

```python
df['Market'].unique()
```

```python
df['Market'].value_counts()
```

```python
price_max=df['Avg'].max()
df[df['Avg']==price_max]
```

```python
price_low=df['Avg'].min()
df[df['Avg']==price_low]
```

```python
df.groupby('Market').mean(numeric_only=True)
```

```python
lst=df['Market'].unique()
lst
```

```python
b_df=df[df['Market']==lst[0]]
c_df=df[df['Market']==lst[1]]
d_df=df[df['Market']==lst[2]]
k_df=df[df['Market']==lst[3]]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14,5))
plt.plot(b_df['Date'],b_df['Avg'],label=lst[0])
plt.plot(c_df['Date'],c_df['Avg'],label=lst[1])
plt.plot(d_df['Date'],d_df['Avg'],label=lst[2])
plt.plot(k_df['Date'],k_df['Avg'],label=lst[3])
plt.legend()
plt.show()
```

```python
import prophet
```

```python
b_df=b_df.rename(columns={'Date':'ds',"Avg":'y'})
b_df=b_df.drop(columns='Market')
b_df
```

```python
from prophet import Prophet
m = Prophet()
m.fit(b_df)
```

```python
future = m.make_future_dataframe(periods=35)
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = m.plot(forecast)
```
